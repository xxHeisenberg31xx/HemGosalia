```python
m = x_train.shape[0]
n = x_train.shape[1]
m, n
```

```python
beta = np.ones(n)
np.reshape(beta, (36,1))
```

```python
def cost(x, y, w):
    
    cost = (1/(2*m))*np.transpose((x@w - y))@(x@w - y)
    return cost
```

```python
a = 0.5
cost_list = []
for i in range(1000000):
    
    beta = beta - a*(1/m)*np.transpose(x_train)@(x_train@beta - y_train)
           
    cost_val = cost(x_train, y_train, beta)
    cost_list.append(cost_val)

cost_list
```
